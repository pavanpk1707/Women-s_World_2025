```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bets_tracker.csv")
```

```python
# Convert 'Profit/Loss' to numeric
df['Profit/Loss'] = pd.to_numeric(df['Profit/Loss'], errors='coerce')

# Convert 'Tournament' to categorical
df['Tournment'] = pd.Categorical(df['Tournment'])

# Print data types after conversion
print(df.dtypes)
```

```python
# Calculate the Total Money column
total_money = []
running_total = 0

for index, row in df.iterrows():
    if row['Won/ Lost'] == 'Won':
        running_total += row['Profit/Loss']
    else:
        running_total -= abs(row['Profit/Loss'])
    total_money.append(running_total)

# Add the Total Money column to the DataFrame
df['Total Money'] = total_money
```

```python
df
```

```python
# Group by tournament and calculate the sum of profit/loss
grouped_df = df.groupby('Tournment')['Profit/Loss'].sum()

print(grouped_df)
```

```python
# Plot the bar graph
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', color='skyblue')
plt.title('Total Profit/Loss by Tournament')
plt.xlabel('Tournament')
plt.ylabel('Total Profit/Loss')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Convert date columns to datetime objects
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Settled date'] = pd.to_datetime(df['Settled date'])

# Sort DataFrame by start time
df.sort_values(by='Start Time', inplace=True)

# Calculate cumulative sum for Profit/Loss
df['Profit_Loss_Cumsum'] = df['Profit/Loss'].cumsum()

# Plotting the time series
plt.figure(figsize=(10, 6))
plt.plot(df['Settled date'], df['Profit_Loss_Cumsum'], label='Total Profit')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.title('Time Series of Total Profit')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 2: Calculate the cumulative profit/loss
df['Cumulative Profit/Loss'] = df['Profit/Loss'].cumsum()

# Step 3: Plot the cumulative profit/loss
plt.figure(figsize=(10, 6))
plt.plot(df['Settled date'], df['Cumulative Profit/Loss'], marker='o', linestyle='-', color='b')
plt.xlabel('Date')
plt.ylabel('Cumulative Profit/Loss')
plt.title('Cumulative Profit/Loss Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
